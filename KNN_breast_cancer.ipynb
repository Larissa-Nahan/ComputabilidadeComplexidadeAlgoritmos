```python
# import das bibliotecas de ciÃªncia de dados
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# carregando o banco de dados (breast_cancer)
from sklearn.datasets import load_breast_cancer
```

```python
# criar novo dataframe e selecionar os valores e dados de alvo target
cancer = load_breast_cancer()

df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
df['class'] = cancer.target
df.head(10)
```

```python
df.describe()
```

```python
x = cancer.data
y = cancer.target.reshape(-1, 1)

print(x.shape, y.shape)
```

```python
# separando base de dados em 70% para treinamento do modelo e 30% para teste e validaÃ§Ã£o
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# metricas de distancia
# tecnica de manhatta
def l1_distance(a, b):
  return np.sum(np.abs(a - b), axis=1)

# tecnica de distancia euclidiana
def l2_distance(a, b):
  return np.sqrt(np.sum((a - b)**2, axis=1))
```

**Implementar o classificados KNN na mÃ£o utilizando a linguagem python**

```python
class kNearestNeighbor(object):
    def __init__(self, n_neighbors=1, dist_func=l1_distance):
        self.n_neighbors = n_neighbors
        self.dist_func = dist_func

    def fit(self, x, y):
        self.x_train = x
        self.y_train = y

    def predict(self, x):
        y_pred = np.zeros((x.shape[0], 1), dtype=self.y_train.dtype)

        for i, x_test in enumerate(x):
            distances = self.dist_func(self.x_train, x_test)
            nn_index = np.argsort(distances)
            nn_pred = self.y_train[nn_index[:self.n_neighbors]].ravel()
            y_pred[i] = np.argmax(np.bincount(nn_pred))

        return y_pred
```

 **Realizando testes e validaÃ§Ãµes com acurÃ¡cia do modelo de machine learning**

```python
knn = kNearestNeighbor(n_neighbors=3)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)

print('AcurÃ¡cia: {:.2f}%'.format(accuracy_score(y_test, y_pred)*100))
```

```python
# Comparativo entre os mÃ©todos do knn - distÃ¢ncia euclidiana e distÃ¢ncia de manhattan
knn = kNearestNeighbor()
knn.fit(x_train, y_train)

list_res = []
for p in [1, 2]:
    knn.dist_func = l1_distance if p == 1 else l2_distance   
    
    for k in range(1, 10, 2):
        knn.n_neighbors = k
        y_pred = knn.predict(x_test)
        acc = accuracy_score(y_test, y_pred)*100
        list_res.append([k, 'l1_distance' if p == 1 else 'l2_distance', acc])
        
df = pd.DataFrame(list_res, columns=['k', 'dist. func.', 'acurÃ¡cia'])
df
```

```python
# comparaÃ§Ã£o com o sklearn
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, p=2)
knn.fit(x_train, y_train.ravel())

list_res = []
for p in [1, 2]:
    knn.p = p
    
    for k in range(1, 10, 2):
        knn.n_neighbors = k
        y_pred = knn.predict(x_test)
        acc = accuracy_score(y_test, y_pred)*100
        list_res.append([k, 'l1_distance' if p == 1 else 'l2_distance', acc])
        
df = pd.DataFrame(list_res, columns=['k', 'dist. func.', 'acurÃ¡cia'])
df
```
